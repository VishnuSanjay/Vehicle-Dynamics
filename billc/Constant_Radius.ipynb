```python
import pandas as pd
import numpy as np
from control.matlab import *
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#Read in the data
steer = pd.read_excel('constantradius.xlsx', sheet_name = "Steer (deg)")
speed = pd.read_excel('constantradius.xlsx', sheet_name = "Speed (kph)")
yawv = pd.read_excel('constantradius.xlsx', sheet_name = "Yaw Velocity (deg per sec)")
roll = pd.read_excel('constantradius.xlsx', sheet_name = "Roll (deg)")
latacc = pd.read_excel('constantradius.xlsx', sheet_name = "Latacc (g)")
sideslip = pd.read_excel('constantradius.xlsx', sheet_name = "SideSlip (deg)")
```

```python
#Make them numpy arrays
steer = np.array(steer)
speed = np.array(speed)/3.6
yawv = np.array(yawv)
roll = np.array(roll)
latacc = np.array(latacc)
sideslip = np.array(sideslip)

#Transpose them, easier to think of time series this way
steer = steer.T
speed = speed.T
yawv = yawv.T
roll = roll.T
latacc = latacc.T
sideslip = sideslip.T

#Car data
frontw = 90 # in kilos
rearw = 110 
wheelbase = 1750 # mm
sr = 6
timestep = 0.001 # seconds

steer_rwa = steer/sr # correct for steer ratio
```

```python
fig, axs = plt.subplots(6, 1)
fig.set_size_inches(18.5, 10.5)
a = 0
axs[0].plot(latacc[a,:])
axs[0].set_xlabel("Sample")
axs[0].set_ylabel("Lateral acceleration (g)")
axs[0].grid()

axs[1].plot(steer[a,:])
axs[1].set_xlabel("Sample")
axs[1].set_ylabel("Steer (deg)")
axs[1].grid()

axs[2].plot(sideslip[a,:])
axs[2].set_xlabel("Sample")
axs[2].set_ylabel("Sideslip (deg)")
axs[2].grid()

axs[3].plot(yawv[a,:])
axs[3].set_xlabel("Sample")
axs[3].set_ylabel("Yaw velocity (deg/sec)")
axs[3].grid()

axs[4].plot(roll[a,:])
axs[4].set_xlabel("Sample")
axs[4].set_ylabel("Roll (deg)")
axs[4].grid()

axs[5].plot(speed[a,:])
axs[5].set_xlabel("Sample")
axs[5].set_ylabel("Speed (m/s)")
axs[5].grid()
```

```python
#Take last column as steady state, assumes steady state achieved
threshold = 300
ss_steer_rwa = np.mean(steer_rwa[:,threshold:-1],axis=1)
ss_steer = np.mean(steer[:,threshold:-1],axis=1)
ss_yawv = np.mean(yawv[:,threshold:-1],axis=1)
ss_roll = np.mean(roll[:,threshold:-1],axis=1)
ss_latacc = np.mean(latacc[:,threshold:-1],axis=1)
ss_sideslip = np.mean(sideslip[:,threshold:-1],axis=1)
ss_speed = np.mean(speed[:,threshold:-1],axis=1)

fig, axs = plt.subplots(5,1)
fig.set_size_inches(18.5, 10.5)

axs[0].plot(ss_latacc,ss_steer)
axs[0].set_xlabel("Lateral acceleration (g)")
axs[0].set_ylabel("Steer (deg)")
axs[0].grid()

axs[1].plot(ss_latacc,ss_roll)
axs[1].set_xlabel("Lateral acceleration (g)")
axs[1].set_ylabel("Roll (deg)")
axs[1].grid()

axs[2].plot(ss_latacc,ss_sideslip)
axs[2].set_xlabel("Lateral acceleration (g)")
axs[2].set_ylabel("Sideslip (deg)")
axs[2].grid()

axs[3].plot(ss_latacc,ss_yawv)
axs[3].set_xlabel("Lateral acceleration (g)")
axs[3].set_ylabel("Yaw velocity (deg/sec)")
axs[3].grid()


axs[4].plot(ss_latacc,ss_steer_rwa)
axs[4].set_xlabel("Lateral acceleration (g)")
axs[4].set_ylabel("Road Wheel Angle (deg)")
axs[4].grid()
```

## 3. Since the plots are linear, slope can be computed using just two points each. We pick first and last ones in each array

```python
# ##For linear models calculating the slope using first and last point is enough
# gain_steer_ay = (ss_steer[-1]-ss_steer[0])/(ss_latacc[-1]-ss_latacc[0]) 
# gain_yawv_ay = (ss_yawv[-1]-ss_yawv[0])/(ss_latacc[-1]-ss_latacc[0]) 
# gain_roll_ay = (ss_roll[-1]-ss_roll[0])/(ss_latacc[-1]-ss_latacc[0]) 
# gain_sideslip_ay = (ss_sideslip[-1]-ss_sideslip[0])/(ss_latacc[-1]-ss_latacc[0]) 

# print("Steer gain = ", gain_steer_ay, "deg/g")
# print("Yaw velocity gain = ", gain_yawv_ay, "deg/s/g")
# print("Roll gain = ", gain_roll_ay, "deg/g")
# print("Sideslip gain = ", gain_sideslip_ay, "deg/g")


#For non-linear models the slope at all points of the curve are required
gain_steer_rwa_ay = (np.gradient(ss_steer_rwa))/(np.gradient(ss_latacc)) #Understeer 
gain_steer_ay = (np.gradient(ss_steer))/(np.gradient(ss_latacc)) 
gain_yawv_ay = (np.gradient(ss_yawv))/(np.gradient(ss_latacc)) 
gain_roll_ay = (np.gradient(ss_roll))/(np.gradient(ss_latacc)) 
gain_sideslip_ay = (np.gradient(ss_sideslip))/(np.gradient(ss_latacc)) 

fig, axs = plt.subplots(4,1)
fig.set_size_inches(18.5, 15.5)

axs[0].plot(ss_latacc,gain_steer_ay)
axs[0].set_xlabel("Lateral acceleration (g)")
axs[0].set_ylabel("Steer/Lat.acc Gain (deg/g)")
axs[0].set_ylim((0,50))
axs[0].grid()

axs[1].plot(ss_latacc,gain_roll_ay)
axs[1].set_xlabel("Lateral acceleration (g)")
axs[1].set_ylabel("Roll/Lat.acc Gain (deg/g)")
axs[1].set_ylim((0,4))
axs[1].grid()

axs[2].plot(ss_latacc,gain_sideslip_ay)
axs[2].set_xlabel("Lateral acceleration (g)")
axs[2].set_ylabel("Sideslip/Lat.acc Gain (deg/g)")
axs[2].set_ylim((-1.5,1.5))
axs[2].grid()

axs[3].plot(ss_latacc,gain_yawv_ay)
axs[3].set_xlabel("Lateral acceleration (g)")
axs[3].set_ylabel("Yaw velocity/Lat.acc Gain (deg/sec/g)")
axs[3].grid()
```

## 3. Tangent Speed Estimation 

```python
sideslip_zero = np.argmin(np.abs(ss_sideslip)) 
Tangent_speed = ss_speed[sideslip_zero]
print('Tangent Speed is =',Tangent_speed,'m/s')

plt.plot(ss_speed,ss_sideslip)
plt.xlabel("Speed (m/s)")
plt.ylabel("Steady State Sideslip (deg)")
plt.grid()

A = rearw / (frontw + rearw) * wheelbase
B = frontw / (frontw + rearw) * wheelbase
DR = -9.81 * 57.3 * B / Tangent_speed**2 / 1000
K = np.mean(gain_steer_rwa_ay)
DF = K - DR
print('Front Cornering Compliance = ',DF,'deg/g')
print('Rear Cornering Compliance = ',DR,'deg/g')
print('Understeer = ',K,'deg/g')
```
